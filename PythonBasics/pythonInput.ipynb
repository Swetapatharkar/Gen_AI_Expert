Check if given number is prime or no

```python

# Take input from the user
num = int(input("Enter a number: "))

# Prime check
if num <= 1:
    print(num, "is not prime")
else:
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            print(num, "is not prime")
            break
    else:
        print(num, "is prime")
```
