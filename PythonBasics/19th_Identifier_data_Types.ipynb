```python
import sys
a=10
print(a)
print(sys.getsizeof(a))  #sys.getsizeof() returns the memory size (in bytes) of a Python object.
print(type(a))
print(a, " is Integer?", isinstance(a,int))
print(sys.getsizeof(12+2j))
```

```python
print(sys.getsizeof(10))        # int
print(sys.getsizeof(10.5))      # float
print(sys.getsizeof("hello"))   # string
print(sys.getsizeof([1,2,3]))  # # Size of list container only.It does not include the memory used by the elements inside the list.
```

```python
print(sys.getsizeof(int()))
print(sys.getsizeof(float()))
print(sys.getsizeof(complex()))
print(sys.getsizeof(str()))
```
