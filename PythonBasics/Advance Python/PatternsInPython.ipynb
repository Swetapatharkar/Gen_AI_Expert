```python
n = 5

# Top half
for i in range(1, n+1):
    for j in range(n-i):
        print(' ', end=' ')
    for j in range(2*i-1):
        if j == 0 or j == 2*i-2:
            print('*', end=' ')
        else:
            print(' ', end=' ')
    print()

# Bottom half
for i in range(n-1, 0, -1):
    for j in range(n-i):
        print(' ', end=' ')
    for j in range(2*i-1):
        if j == 0 or j == 2*i-2:
            print('*', end=' ')
        else:
            print(' ', end=' ')
    print()
```

```python
n = 5

# Top half
for i in range(1, n+1):
    for j in range(n-i):
        print(" ", end=" ")
    
    # increasing numbers
    for j in range(1, i+1):
        print(j, end=" ")
    
    # decreasing numbers
    for j in range(i-1, 0, -1):
        print(j, end=" ")
    
    print()

# Bottom half
for i in range(n-1, 0, -1):
    for j in range(n-i):
        print(" ", end=" ")
    
    for j in range(1, i+1):
        print(j, end=" ")
    
    for j in range(i-1, 0, -1):
        print(j, end=" ")
    
    print()
```

```python
n = 5

# Top half
for i in range(1, n+1):
    for j in range(n-i):
        print(" ", end=" ")
    
    for j in range(1, 2*i):
        if j == 1 or j == 2*i-1:
            print(j if j <= i else 2*i-j, end=" ")
        else:
            print(" ", end=" ")
    
    print()

# Bottom half
for i in range(n-1, 0, -1):
    for j in range(n-i):
        print(" ", end=" ")
    
    for j in range(1, 2*i):
        if j == 1 or j == 2*i-1:
            print(j if j <= i else 2*i-j, end=" ")
        else:
            print(" ", end=" ")
    
    print()
```

```python
n = 5

# Upper half
for i in range(1, n+1):
    # Left numbers
    for j in range(1, i+1):
        print(j, end=" ")
    
    # Spaces
    for j in range(2*(n-i)):
        print(" ", end=" ")
    
    # Right stars
    for j in range(i):
        print("*", end=" ")
    
    print()

# Lower half
for i in range(n-1, 0, -1):
    for j in range(1, i+1):
        print(j, end=" ")
    
    for j in range(2*(n-i)):
        print(" ", end=" ")
    
    for j in range(i):
        print("*", end=" ")
    
    print()
```

```python
n = 5

# -------- TOP (Numbers Butterfly) --------
for i in range(1, n+1):
    # Left numbers
    for j in range(1, i+1):
        print(j, end=" ")
    
    # Spaces
    for j in range(2*(n-i)):
        print(" ", end=" ")
    
    # Right numbers
    for j in range(1, i+1):
        print(j, end=" ")
    
    print()

for i in range(n-1, 0, -1):
    # Left numbers
    for j in range(1, i+1):
        print(j, end=" ")
    
    # Spaces
    for j in range(2*(n-i)):
        print(" ", end=" ")
    
    # Right numbers
    for j in range(1, i+1):
        print(j, end=" ")
    
    print()

# -------- BOTTOM (Star Butterfly) --------
for i in range(1, n+1):
    # Left stars
    for j in range(i):
        print("*", end=" ")
    
    # Spaces
    for j in range(2*(n-i)):
        print(" ", end=" ")
    
    # Right stars
    for j in range(i):
        print("*", end=" ")
    
    print()

for i in range(n-1, 0, -1):
    # Left stars
    for j in range(i):
        print("*", end=" ")
    
    # Spaces
    for j in range(2*(n-i)):
        print(" ", end=" ")
    
    # Right stars
    for j in range(i):
        print("*", end=" ")
    
    print()
```

```python
n = 5

for i in range(1, n+1):
    # spaces for alignment
    for j in range(n-i):
        print(" ", end=" ")
    
    # numbers with hollow logic
    for j in range(1, 2*i):
        if j == 1 or j == 2*i-1 or i == n:
            print(j if j <= i else 2*i-j, end=" ")
        else:
            print(" ", end=" ")
    
    print()
```

```python
n = 5

# 1️⃣ Star Pyramid
for i in range(1, n+1):
    print("* " * i)

# 2️⃣ Number Pyramid
for i in range(1, n+1):
    for j in range(1, i+1):
        print(j, end=" ")
    print()
```

```python
n = 5

# 1️⃣ Star Pyramid (Right-aligned)
for i in range(1, n+1):
    print("  "*(n-i) + "* "*i)

# 2️⃣ Number Pyramid (Right-aligned)
for i in range(1, n+1):
    print("  "*(n-i), end="")
    for j in range(1, i+1):
        print(j, end=" ")
    print()
```
