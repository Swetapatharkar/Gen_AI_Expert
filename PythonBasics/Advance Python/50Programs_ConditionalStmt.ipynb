```python
# Determine if a given date is valid (considering month length and leap year for February). 
day = int(input("Enter day: "))
month = int(input("Enter month: "))
year = int(input("Enter year: "))

valid = True

if month < 1 or month > 12:
    valid = False

elif month == 2:
    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
        if day < 1 or day > 29:
            valid = False
    else:
        if day < 1 or day > 28:
            valid = False

elif month in [4,6,9,11]:
    if day < 1 or day > 30:
        valid = False

else:
    if day < 1 or day > 31:
        valid = False

if valid:
    print("Valid Date")
else:
    print("Invalid Date")
```

```python
#Check if a number is an Armstrong number (sum of its digits raised to the 
#power of the number of digits equals the number). 
num = input("Enter a number: ")
power = len(num)

if sum(int(digit) ** power for digit in num) == int(num):
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

```python
# Determine the type of quadrilateral based on given angles and sides.
s1 = int(input("Enter side 1: "))
s2 = int(input("Enter side 2: "))
s3 = int(input("Enter side 3: "))
s4 = int(input("Enter side 4: "))

a1 = int(input("Enter angle 1: "))
a2 = int(input("Enter angle 2: "))
a3 = int(input("Enter angle 3: "))
a4 = int(input("Enter angle 4: "))

# Check quadrilateral type
if s1 == s2 == s3 == s4 and a1 == a2 == a3 == a4 == 90:
    print("Square")

elif s1 == s3 and s2 == s4 and a1 == a2 == a3 == a4 == 90:
    print("Rectangle")

elif s1 == s2 == s3 == s4:
    print("Rhombus")

elif s1 == s3 and s2 == s4 and a1 == a3 and a2 == a4:
    print("Parallelogram")

else:
    print("Other Quadrilateral")
```

```python
# Write a program that converts a Roman numeral to an integer. 
roman = input("Enter Roman Numeral: ").upper()

roman_dict = {
    'I': 1,
    'V': 5,
    'X': 10,
    'L': 50,
    'C': 100,
    'D': 500,
    'M': 1000
}

total = 0
prev_value = 0

for ch in reversed(roman):
    value = roman_dict[ch]
    
    if value < prev_value:
        total -= value
    else:
        total += value
    
    prev_value = value

print("Integer value:", total)
```

```python
# Implement a rock-paper-scissors game using conditional statements
import random

choices = ["rock", "paper", "scissors"]

user = input("Enter rock, paper, or scissors: ").lower()
computer = random.choice(choices)

print("Computer chose:", computer)

if user == computer:
    print("It's a tie!")

elif (user == "rock" and computer == "scissors") or \
     (user == "scissors" and computer == "paper") or \
     (user == "paper" and computer == "rock"):
    
    print("You win!")

else:
    print("You lose!") 
```

```python
# Write a program to find the day of the week for a given date (without usingn built-in functions). 
def day_of_week(d, m, y):
    if m < 3:
        m += 12
        y -= 1

    k = y % 100
    j = y // 100

    day = (d + (13 * (m + 1)) // 5 + k + (k // 4) + (j // 4) - 2 * j) % 7

    days = ["Saturday", "Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]

    return days[day]


day, month, year = map(int, input("Enter date (DD MM YYYY): ").split())

print("Day of the Week:", day_of_week(day, month, year))
```
