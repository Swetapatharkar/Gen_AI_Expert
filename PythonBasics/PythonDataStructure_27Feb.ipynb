=====================SET==================
Sets do not allow duplicate values.

Sets are unordered, so elements do not have an index.

Sets are mutable, which means you can add or remove items.

Sets are often used for membership testing, removing duplicates, and mathematical operations like union, intersection, difference.

```python
fruits={'Apple','Mangoes','Watermelon','Jackfruit'}
#print(fruits) #Order of display may change
repeatList={1,2,3,3,4,4,'Apple','Apple','Banana','Banana'}
noRepeatList=set(repeatList)
print(noRepeatList)
```
