############ When we will get RGB intensity images ############
They do not give colored images by themselves.
To see colored images, you must keep the RGB triplets together:
    plt.imshow(ele_red[:,:,0]), you see a grayscale image of the red intensity.
    plt.imshow(ele_red[:,:,1]), you see a grayscale image of the green intensity.
    plt.imshow(ele_red[:,:,2]), you see a grayscale image of the blue intensity.

```python
# Red Image
red_img = ele_array.copy()
red_img[:,:,1] = 0
red_img[:,:,2] = 0
plt.imshow(red_img)
```

```python
# Green Image
green_img = ele_array.copy()
green_img[:,:,0] = 0
green_img[:,:,2] = 0
plt.imshow(green_img)
```

```python
# Blue Image
blue_img = ele_array.copy()
blue_img[:,:,0] = 0
blue_img[:,:,1] = 0
plt.imshow(blue_img)
```

############ plaot the all images in a row  ############

```python
plt.figure(figsize=(16,4))  # width x height in inches
```

```python
# Original image
plt.subplot(1,4,1)
plt.imshow(ele_array)
plt.title("Original")
plt.axis("off")
```

```python
# Red image
plt.subplot(1,4,2)
plt.imshow(red_img)
plt.title("Red")
plt.axis("off")
```

```python
# Green image
plt.subplot(1,4,3)
plt.imshow(green_img)
plt.title("Green")
plt.axis("off")
```

```python
# Blue image
plt.subplot(1,4,4)
plt.imshow(blue_img)
plt.title("Blue")
plt.axis("off")
```

################### PRINT ALL THE IMAGES: ORIGINAL,RED,GREEN AND BLUE IN ONE FRAME

```python
# Create a figure with 1 row, 4 columns
plt.figure(figsize=(16,4))  # width x height in inches

# Original image
plt.subplot(1,4,1)
plt.imshow(ele_array)
plt.title("Original")
plt.axis("off")

# Red image
plt.subplot(1,4,2)
plt.imshow(red_img)
plt.title("Red")
plt.axis("off")

# Green image
plt.subplot(1,4,3)
plt.imshow(green_img)
plt.title("Green")
plt.axis("off")

# Blue image
plt.subplot(1,4,4)
plt.imshow(blue_img)
plt.title("Blue")
plt.axis("off")

plt.tight_layout()
plt.show()
```
