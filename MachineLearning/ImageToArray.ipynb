====================================================
Code	Result
[:,:,0]	Red channel
[:,:,1]	Green channel
[:,:,2]	Blue channel
[:,:,1] = 0	Remove green
[:,:,2] = 0	Remove blue
both = 0	Only red remains

```python
plt.imshow(ele_red[:,:,0], cmap='gray')  # Red
plt.title("Red channel")

plt.imshow(ele_red[:,:,2], cmap='gray')  # Blue
plt.title("Blue channel")

# to restore green channel
ele_red[:,:,1] = ele_array[:,:,1]

plt.imshow(ele_red[:,:,1], cmap='gray')  # Green
plt.title("Green channel")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))

# Original image
plt.subplot(1,4,1)
plt.imshow(ele_red)
plt.title("Original")
plt.axis('off')

# Red channel
plt.subplot(1,4,2)
plt.imshow(ele_red[:,:,0], cmap='gray')
plt.title("Red Channel")
plt.axis('off')

# Green channel
plt.subplot(1,4,3)
plt.imshow(ele_red[:,:,1], cmap='gray')
#plt.imshow(ele_array[:,:,1], cmap='gray') # used this code to restore green array to normal as it was set to 0 earlier
plt.title("Green channel")
plt.axis('off')

# Blue channel
plt.subplot(1,4,4)
plt.imshow(ele_red[:,:,2], cmap='gray')
plt.title("Blue Channel")
plt.axis('off')

plt.show()
```
