============== Stacking & Splitting Arrays =================
np.hstack()	horizontal join
np.vstack()	vertical join
np.concatenate() general join

hstack → horizontal → side by side
vstack → vertical → top and bottom

hstack → horizontal → rows must match
vstack → vertical → columns must match

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
```

```python
hstack=np.hstack([a,b])  # Rows must match, NumPy places b to the right of a.
hstack 
```

```python
vstack=np.vstack([a,b]) # NumPy places b below a.
vstack
```

```python
p=np.array([1,2])
q=np.array([5,6])
print(np.vstack([p,q]))
```

================= Boolean Masking & Fancy Indexing ============
