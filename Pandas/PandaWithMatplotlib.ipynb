```python
#Example visualization
#Internet users vs Birth rate (scatter plot)
import matplotlib.pyplot as plt

plt.scatter(df['InternetUsers'], df['BirthRate']) # Scatter Plot is Used to find relationships between variables.

plt.xlabel("Internet Users (%)")
plt.ylabel("Birth Rate")
plt.title("Birth Rate vs Internet Usage")

plt.show()
```

```python
#Birth rate by income group:
import seaborn as sns

sns.boxplot(x="IncomeGroup", y="BirthRate", data=df,palette="Set2") # Box plot Shows median, quartiles, and outliers.
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x="IncomeGroup", 
            y="BirthRate", 
            hue="IncomeGroup", 
            data=df, 
            palette="Set2",
            legend=True)

plt.show()
```

```python
sns.boxplot(x="IncomeGroup", y="BirthRate", data=df, hue="IncomeGroup",palette="Set2")

plt.title("Birth Rate by Income Group")
plt.show()
```

```python
# KDE Plot (Smooth Distribution)
sns.kdeplot(df["BirthRate"], fill=True)

plt.show()
#Used to understand probability density.
```

```python
#Pair Plot (VERY POWERFUL)
sns.pairplot(df, hue="IncomeGroup")

plt.show()
#Creates multiple scatter plots automatically.
#Used in:
#exploratory data analysis
#feature relationship discovery
```

```python
# Heatmap (Correlation Matrix)
import pandas as pd

corr = df[["BirthRate","InternetUsers"]].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.show()
#Shows correlation between variables.
```

```python
#Violin Plot (Better than Boxplot)
sns.violinplot(
    x="IncomeGroup",
    y="BirthRate",
    data=df
)

plt.show()
#Shows:
#distribution
#density
#quartiles
```

```python
#Count Plot
sns.countplot(x="IncomeGroup", data=df)

plt.show()
#Shows how many countries belong to each income group.
```

```python
#Regression Plot (VERY useful in ML)
sns.regplot(
    x="InternetUsers",
    y="BirthRate",
    data=df
)

plt.show()
#Shows:
#scatter
#best fit line
#Used to detect linear relationships.
```

```python
#Joint Plot
sns.jointplot(
    x="InternetUsers",
    y="BirthRate",
    data=df,
    kind="scatter"
)

plt.show()
#Shows
#scatter plot
#two histograms
```
